```python
# mtcars.csv
```

```python
import pandas as pd
```

```python
df = pd.read_csv('mtcars.csv')
df.shape
```

```python
df.columns
```

```python
x = df[['disp', 'hp', 'wt']]
y = df['mpg']
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x, y)
```

```python
reg.score(x, y)
```

# recursive feature elimination

```python
from sklearn.feature_selection import RFE
```

```python
rfe = RFE(reg)
```

```python
x = df.drop('mpg', axis = 1)
```

```python
x.columns
```

```python
rfe.fit(x, y)
```

```python
rfe.get_support()
```

```python
x = x.iloc[:,rfe.get_support()]
```

```python
x
```

```python
reg.fit(x, y)
```

```python
reg.score(x, y)
```

# forward/ backward

```python
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
sfs = SequentialFeatureSelector(reg, direction= 'backward')
```

```python
x = df.drop('mpg', axis = 1)
```

```python
sfs.fit(x, y)
```

```python
sfs.get_support()
```

```python
x = x.iloc[:,sfs.get_support()]
x.head()
```

```python
reg.fit(x, y)
reg.score(x, y)
```
