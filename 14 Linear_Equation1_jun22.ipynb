```python
# input data-> height in cm
x = [[151],[174],[138],[186],[128],[136],[179],[163],[152],[131]]
```

```python
# output data in kg
y = [63, 81, 56, 91, 47, 57, 76, 72, 62, 48]
```

```python
# import the class
from sklearn.linear_model import LinearRegression
```

```python
regressor.fit(x, y)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```
