```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df
```

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.scatter(df['Age'], df['EstimatedSalary'], c = df['Purchased']);
```

### Normalization

```python
# import the class
from sklearn.preprocessing import MinMaxScaler
```

```python
# create an object
scaler = MinMaxScaler()
```

```python
x = df[['Age', 'EstimatedSalary']]
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
x_scaled[:,0];
```

```python
x_scaled[:,1];
```

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.scatter(x_scaled[:,0], x_scaled[:,1], c = df['Purchased']);
```

```python
x_scaled[:,0].min(), x_scaled[:,0].max()
```

```python
x_scaled[:,1].min(), x_scaled[:,1].max()
```

### Standardization

```python
# import the class
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```
