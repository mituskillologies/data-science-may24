```python
# separate input and output data

x = df[['Age', 'EstimatedSalary']]

y = df['Purchased']
```

### feature scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
```

```python
pd.DataFrame(x_scaled).describe()
```

#### cross validation

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y,
                                                   random_state= 0, test_size= 0.25)
```

```python
x_train.shape
```

```python
y_train.shape
```

### Build the model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
clf_log = LogisticRegression()
clf_nb = GaussianNB()
clf_dt = DecisionTreeClassifier(random_state= 0)
clf_svm = SVC(random_state= 0)
clf_knn = KNeighborsClassifier(n_neighbors= 5)
```

```python
clf_log.fit(x_train, y_train)
clf_nb.fit(x_train, y_train)
clf_dt.fit(x_train, y_train)
clf_svm.fit(x_train, y_train)
clf_knn.fit(x_train, y_train)
```

### prediction on test data

```python
y_pred_log = clf_log.predict(x_test)
y_pred_nb = clf_nb.predict(x_test)
y_pred_dt = clf_dt.predict(x_test)
y_pred_svm = clf_svm.predict(x_test)
y_pred_knn = clf_knn.predict(x_test)
```

#### Performance Evaluation

```python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
print('acc log', accuracy_score(y_test, y_pred_log))
print('acc nb', accuracy_score(y_test, y_pred_nb))
print('acc dt', accuracy_score(y_test, y_pred_dt))
print('acc svm', accuracy_score(y_test, y_pred_svm))
print('acc knn', accuracy_score(y_test, y_pred_knn))
```
