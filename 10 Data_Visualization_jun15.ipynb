# Boxplot

```python
sm = [67, 62, 70, 63, 66, 65, 71, 61, 61, 85]
```

```python
plt.boxplot(sm);
```

```python
q1 = np.percentile(sm, 25)
q3 = np.percentile(sm, 75)
print(q1, q3)
```

```python
iqr = q3 - q1
iqr
```

```python
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
print(lower_bound, upper_bound)
```

```python
# function for outlier

def outlier(data):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)

    iqr = q3 - q1

    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)

    outl = [x for x in data if x < lower_bound or x > upper_bound]

    return outl
```

```python
outlier(sm)
```
